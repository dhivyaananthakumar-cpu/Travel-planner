```python
data = {
    "Distance_km": [100, 250, 500, 150],
    "Transport": ["Bus", "Car", "Flight", "Train"],
    "CO2_Emission": [10, 40, 150, 15]
}

df = pd.DataFrame(data)
```

```python
best_option = df.loc[df["CO2_Emission"].idxmin()]

print("Eco-Friendly Travel Recommendation")
print("----------------------------------")
print("Transport:", best_option["Transport"])
print("Distance:", best_option["Distance_km"], "km")
print("CO2 Emission:", best_option["CO2_Emission"], "kg")
```
